```python
import hiragana_tokenizer as hir
```

```python
toker = hir.HiraganaTokenizer()
toker.vocab_size, len(toker.tokens), len(toker.stoi), len({s:i for i,s in enumerate(toker.tokens)})
```

```python
# looks like multiple symbols have the same romanization :o - FIXED
print(len(set(toker.tokens)), len(toker.tokens))
seen = []
for i,t in enumerate(toker.tokens):
    if t not in seen:
        seen.append(t)
    else:
        print(f"Duplicate token: {t} at index {i}")
```

```python
# import data:
with open('Data/samurai_names.txt', 'r') as d:
    nams = d.read().lower()#.splitlines()
names = [' '.join(n.lower().split(" ")) + "\n" for n in nams.splitlines()]
#names, len(names)
```

```python
num = 0
enc = []
for n in names: 
    enc_n = toker(n)
    enc += enc_n
    if len(enc_n):
        num += 1
print(f'Successfully encoded names: {num}')
```

```python
dec_names = toker.decode(enc)
print(dec_names)
```

```python
toker.reduceTokens(enc)
```

```python
rnum = 0
renc = []
for n in names: 
    renc_n = toker(n)
    renc += renc_n
    if len(renc_n):
        rnum += 1
print(f'Successfully encoded names: {rnum}')
```

```python
for n in names:
    print(toker.decode(toker(n)))
```
